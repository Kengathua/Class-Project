```python
def save_historical_data():
    if not os.path.exists('historical_nse_dfs'):
            os.makedirs('historical_nse_dfs')

    fieldnames = ['Date', 'Open', 'Low', 'High', 'Close', 'Adj Close', 'Volume']

    for count, stock in enumerate(stocks):
        ticker = stock[2]
        date=stock[0]
        print(ticker)

        try:
            try:
                low = float(stock[4])
                high = float(stock[5])
                close = float(stock[6])
                start = float(stock[7])
                volume = float(stock[8])

            except ValueError:
                low = stock[4].replace('-', "0")
                high = stock[5].replace('-', "0")
                close = stock[6].replace('-', "0")
                start = stock[7].replace('-', "0")
                volume = stock[8].replace('-', "0")

        except ValueError:
            low = float(stock[4].replace(',', ""))
            high = float(stock[5].replace(',', ""))
            close = float(stock[6].replace(',', ""))
            start = float(stock[7].replace(',', ""))
            volume = float(stock[8].replace(',', ""))
        
        # print(date, ticker, low, high, close, start, volume)

        rows = [{
                'Date': date,
                'Open': start,
                'Low': low,
                'High': high,
                'Close': close,
                'Adj Close': close,
                'Volume': volume,
            }]

        if os.path.exists("historical_nse_dfs/{}.csv".format(ticker)):
            with open('historical_nse_dfs/{}.csv'.format(ticker), "a", encoding='UTF8', newline='') as f:
                    print('Appending')
                    csv_writer = csv.DictWriter(
                        f, fieldnames=fieldnames)
                    csv_writer.writerows(rows)

        else:
            with open("historical_nse_dfs/{}.csv".format(ticker), "w", encoding='UTF8', newline='') as f:
                    print('Writing')
                    csv_writer = csv.DictWriter(
                        f, fieldnames=fieldnames)
                    csv_writer.writeheader()
                    csv_writer.writerows(rows)
        
        if count % 10 == 0:
            print(count)

save_historical_data()
```
