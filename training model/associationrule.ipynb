```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset=[
    ['Milk','Onion','Bread','Kidney Beans','Eggs','Yoghurt'],
    ['Fish','Onion','Bread','Kidney Beans','Eggs', 'Yoghurt'],
    ['Milk','Apple','Kidney Beans','Eggs'],
    ['Milk','Sugar','Tea Leaves','Kidney Beans','Yoghurt'],
    ['Tea Leaves','Onion','Kidney Beans','Ice Cream','Eggs']]
```

```python
tr=TransactionEncoder()
tr_arr = tr.fit(dataset).transform(dataset)
```

```python
df=pd.DataFrame(tr_arr, columns=tr.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
res
```

```python
res1 = res[['antecedents','consequents','support','confidence','lift']]
res1
```

```python
res2=res1[res1['confidence']>=1]
res2
```
